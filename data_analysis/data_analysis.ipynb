```python
import os
import subprocess

# Specify the relative path to trec_eval executable
trec_eval_path = os.path.join('..', 'IR2024', 'trec_eval', 'trec_eval.exe')

# Define the range of parameters for BM25 and LMJelinekMercer
k1_values = [0.5, 1.0, 1.5, 2.0]
b_values = [0.0, 0.25, 0.5, 0.75, 1.0]
lambda_values = [0.1, 0.3, 0.5, 0.7, 0.9]

# Paths to the qrels file and the results directory
qrels_file = '../IR2024/trec_eval/qrels.txt'
results_dir = '../IR2024/trec_eval/results/'
experiments_dir = '../IR2024/trec_eval/experiments/'

# Ensure the results directory exists
os.makedirs(results_dir, exist_ok=True)

# Check if qrels file exists
if not os.path.exists(qrels_file):
    print(f'Error: qrels file not found at {qrels_file}')
    exit(1)

# Function to run trec_eval and save the results
def run_trec_eval(result_file, output_file_prefix):
    metrics = ['map'] + [f'P.{k}' for k in [5, 10, 15, 20]]
    for metric in metrics:
        command = f'trec_eval -q -m {metric} {qrels_file} {result_file}'
        output_file = f'{results_dir}{output_file_prefix}_{metric}.txt'
        
        # Print and execute the command
        print(f'Running command: {command}')
        
        # Run the command and capture the output
        process = subprocess.run(command, shell=True, capture_output=True, text=True)
        
        # Write command output to the file
        with open(output_file, 'w') as f:
            f.write(process.stdout)
        
        # Print any errors
        if process.stderr:
            print(f'Error: {process.stderr}')

# Run trec_eval for BM25
for k1 in k1_values:
    for b in b_values:
        result_file_prefix = f'BM25_k1_{k1}_b_{b}'
        result_file = f'{experiments_dir}{result_file_prefix}.txt'
        
        # Check if result file exists
        if not os.path.exists(result_file):
            print(f'Error: result file not found at {result_file}')
            continue
        
        run_trec_eval(result_file, result_file_prefix)

# Run trec_eval for LMJelinekMercer
for lambda_val in lambda_values:
    result_file_prefix = f'LMJM_lambda_{lambda_val}'
    result_file = f'{experiments_dir}{result_file_prefix}.txt'
    
    # Check if result file exists
    if not os.path.exists(result_file):
        print(f'Error: result file not found at {result_file}')
        continue
    
    run_trec_eval(result_file, result_file_prefix)
```

```python
# Specify the relative path to trec_eval executable
trec_eval_path = os.path.join('..', 'IR2024', 'trec_eval', 'trec_eval.exe')

# Define the range of parameters for BM25 and LMJelinekMercer
k1_values = [0.5, 1.0, 1.5, 2.0]
b_values = [0.0, 0.25, 0.5, 0.75, 1.0]
lambda_values = [0.1, 0.3, 0.5, 0.7, 0.9]

# Paths to the qrels file and the results directory
qrels_file = '../IR2024/trec_eval/qrels.txt'
results_dir = '../IR2024/trec_eval/results/'
experiments_dir = '../IR2024/trec_eval/experiments/'

# Ensure the results directory exists
os.makedirs(results_dir, exist_ok=True)

# Function to run trec_eval and save the results
def run_trec_eval(result_file, output_file_prefix):
    metrics = ['map'] + [f'P.{k}' for k in [5, 10, 15, 20]]
    for metric in metrics:
        command = f'"{trec_eval_path}" -q -m {metric} {qrels_file} {result_file}'
        output_file = f'{results_dir}{output_file_prefix}_{metric}.txt'
        process = subprocess.run(command, shell=True, capture_output=True, text=True)
        with open(output_file, 'w') as f:
            f.write(process.stdout)


# Run trec_eval for BM25
for k1 in k1_values:
    for b in b_values:
        result_file_prefix = f'BM25_k1_{k1}_b_{b}'
        result_file = f'{experiments_dir}{result_file_prefix}.txt'
        run_trec_eval(result_file, result_file_prefix)

# Run trec_eval for LMJelinekMercer
for lambda_val in lambda_values:
    result_file_prefix = f'LMJM_lambda_{lambda_val}'
    result_file = f'{experiments_dir}{result_file_prefix}.txt'
    run_trec_eval(result_file, result_file_prefix)
```

```python
# Function to load results from trec_eval output files
def load_results(result_dir, metric, prefixes):
    results = {}
    for prefix in prefixes:
        file = f'{result_dir}{prefix}_{metric}.txt'
        with open(file) as f:
            data = {}
            for line in f:
                parts = line.split()
                if parts[1].startswith("Q"):
                    query_id = parts[1]
                    value = float(parts[2])
                    data[query_id] = value
            results[prefix] = data
    return results

# Load results
# MAP metric
map_results_bm25 = load_results(results_dir, 'map', bm25_prefixes)
map_results_lmjm = load_results(results_dir, 'map', lmjm_prefixes)
```

```python
# Function to plot results
def plot_results(results, x_values, x_label, metric):
    fig, ax = plt.subplots()
    for label, data in results.items():
        y_values = [sum(data.values())/len(data) for x in x_values]
        ax.plot(x_values, y_values, label=label)
    ax.set_xlabel(x_label)
    ax.set_ylabel(metric)
    ax.legend()
    plt.title(f'{metric} vs {x_label}')
    plt.show()
```
