- Let's start by importing the necessary libraries

```python
import os
import subprocess

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

- Firstly, we will specify the relative path to the `trec_eval.exe`, which will be used for evaluating the results obtained from running the models (in `Phase2.java`).
- Next, we will define the range of parameters for the `BM25` and `LMJelinekMercer` models.
    - k1 and b values for BM25
    - lambda values for LMJelinekMercer.
- We will specify the paths to the qrels file (containing ground truth relevance judgments) and the directories for storing results and experiments.
- A function named `run_trec_eval()` will be defined to execute `trec_eval` with specified parameters and save the results.
- We will run `trec_eval` for various combinations of k1 and b values for `BM25`, and save the results.
- Similarly, we will run it for different lambda values for `LMJelinekMercer`.

```python
# Specify the relative path to trec_eval executable
trec_eval_path = os.path.join('..', 'IR2024', 'trec_eval', 'trec_eval.exe')

# Define the range of parameters for BM25 and LMJelinekMercer
k1_values = [0.5, 1.0, 1.5, 2.0]
b_values = [0.0, 0.25, 0.5, 0.75, 1.0]
lambda_values = [0.0, 0.1, 0.3, 0.5, 0.7, 0.9, 1.0]

# Paths to the qrels file and the results directory
qrels_file = '../IR2024/trec_eval/qrels.txt'
results_dir = '../IR2024/trec_eval/results/'
experiments_dir = '../IR2024/trec_eval/experiments/'

# Ensure the results directory exists
os.makedirs(results_dir, exist_ok=True)

# Function to run trec_eval and save the results
def run_trec_eval(result_file, output_file_prefix):
    metrics = ['map'] + [f'P.{k}' for k in [5, 10, 15, 20]]
    for metric in metrics:
        command = f'"{trec_eval_path}" -q -m {metric} {qrels_file} {result_file}'
        output_file = f'{results_dir}{output_file_prefix}_{metric}.txt'
        
        # Print and execute the command and capture the output
        print(f'Running command: {command}')
        process = subprocess.run(command, shell=True, capture_output=True, text=True)
        
        with open(output_file, 'w') as f:
            f.write(process.stdout)

# Run trec_eval for BM25
for k1 in k1_values:
    for b in b_values:
        result_file_prefix = f'BM25_k1_{k1}_b_{b}'
        result_file = f'{experiments_dir}{result_file_prefix}.txt'
        run_trec_eval(result_file, result_file_prefix)

# Run trec_eval for LMJelinekMercer
for lambda_val in lambda_values:
    result_file_prefix = f'LMJM_lambda_{lambda_val}'
    result_file = f'{experiments_dir}{result_file_prefix}.txt'
    run_trec_eval(result_file, result_file_prefix)
```

```python
# Function to load results from trec_eval output files
def load_results(result_dir, metric, prefixes):
    results = {}
    for prefix in prefixes:
        file = f'{result_dir}{prefix}_{metric}.txt'
        with open(file) as f:
            data = {}
            for line in f:
                parts = line.split()
                # Check if the line is for a specific query or the "all" line
                if parts[1].startswith("Q") or parts[1] == "all":
                    query_id = parts[1]
                    value = float(parts[2])
                    data[query_id] = value
            results[prefix] = data
    return results
```

## Data Visualization of the MAP metric

```python
# Convert the column headers to strings for plotting
df.columns = [str(col) for col in df.columns]

# Calculate mean map for each (k1, b) configuration
mean_map = df.mean()

# Plot mean precision for each (k1, b) configuration
plt.figure(figsize=(14, 8))
mean_map.plot(kind='bar')
plt.title('Mean MAP across different (k1, b) configurations')
plt.xlabel('(k1, b)')
plt.ylabel('Mean MAP')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot line plots for each query
plt.figure(figsize=(14, 8))
for query in df.index:
    plt.plot(df.columns, df.loc[query], marker='o', label=query)

plt.title('MAP across different (k1, b) configurations')
plt.xlabel('(k1, b)')
plt.ylabel('MAP')

# Define a custom color palette for the legend
legend_palette = sns.color_palette("husl", len(df.columns))
plt.legend(title='Query', loc='best')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(df, annot=True, cmap='viridis')
plt.title('Heatmap of MAP for each Query and (k1, b) configuration')
plt.xlabel('(k1, b)')
plt.ylabel('Query')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Results Explanation

## Outliers
- From the above tables and plots we can see that in both models and for all parameters the **Precision** and **MAP** metrics for queries Q01("multimodal travel services") and Q08("Seamless Efficient European Travelling") are extremely high, whereas those for queries Q04("Architectures for Big Data Analytics") and Q06("Mobility-as-a-Service tools") are very low. This can happen because the performance of BM25Similarity varies significantly depending on the nature of the queries and the document collection. High MAP scores are typically achieved for queries that are specific, clear, and well-represented in the document set. In contrast, low MAP scores occur for queries that are broad, ambiguous, or not well-covered by the available documents. 

- We can also notice that the results we get for the LMJelinekMercerSimilarity for **lambda** values **0.0, 1.0** are very poor.
    - When **lambda is set to 0.0** the model relies entirely on the collection language model, meaning that it ignores the specific content of the document and only considers the overall frequency of words in the collection. As a result, the retrieval becomes very generic and fails to capture the nuances of individual documents, leading to poor relevance and retrieval performance.
    - When **lambda is set to 1.0** the model relies entirely on the document language model, meaning that it only considers the words in the document and ignores the overall distribution of words in the collection. As a result, the model becomes very sensitive to the specifics of individual documents and can overfit to the noise or sparse data within a single document, leading to poor generalization and retrieval performance.
    - **To sum up** the extreme values of lambda (0.0 and 1.0) lead to poor performance because they fail to balance the document-specific information with the collection-wide statistics, resulting in either overly generic or overly specific retrieval. The intermediate values of lambda provide a better balance, leading to improved retrieval performance as they leverage the strengths of both the document model and the collection model.
